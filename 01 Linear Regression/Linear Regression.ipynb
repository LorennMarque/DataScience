# Importar requerimientos

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cleanData.csv') # Cargar dataset, archivo CSV
```

```python
data.describe() # Funcion para una overview del contenido en el CSV, en notacion cientifica.
```

```python
x = DataFrame(data, columns=['prod_budget']) # Guardar los valores de la columna prod_budget
y = DataFrame(data, columns=['revenue'])
x # Mostrar los valores de la columna prod_budget
```

```python
plt.figure(figsize=(10,6)) # Definir el tamaÃ±o del grÃ¡fico, width and height
plt.title("Costo de produccion vs Ganancia Global", color="white", fontsize=16) # Titulo, color y tamaÃ±o
plt.xlabel("Presupuesto $", color="white", fontsize=14) # Etiqueta para el eje X, tamaÃ±o de letra y color
plt.ylabel("Ganancia $", color="white", fontsize=14)
plt.ylim(0,3000000000) # Limitar valores en eje Y
plt.xlim(0,450000000)
plt.scatter(x, y, alpha=0.3) # Generar grÃ¡fico y definir opacidad de puntos
plt.show() # Funcion para mostrar el grÃ¡fico
```

```python
regression = LinearRegression() # Definir una variable para la funcion
regression.fit(x, y)
```

```python
regression.coef_  #theta_1
```

```python
# Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=0.3)

plt.plot(x,regression.predict(x), color="red", linewidth=4) # Crear linea de prediccion, color y ancho.

plt.title("Costo de produccion vs Ganancia Global", color="white", fontsize=16)
plt.xlabel("Presupuesto $", color="white", fontsize=14)
plt.ylabel("Ganancia $", color="white", fontsize=14)
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression.score(x, y)
```
