## Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cost_revenue_clean.csv")
```

```python
data.describe()
```

```python
X = pd.DataFrame(data, columns=["production_budget_usd"]) # Columns
y = pd.DataFrame(data, columns=["worldwide_gross_usd"]) # Target variable
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title("Film cost vs Film Revenue")
plt.xlabel("Production Budget")
plt.ylabel("Revenue U$D")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

### Traducción de algunos terminos importantes:
* La ordenada al origen es "Intercept"
* La pendiente es "Slope"

### Notas
La pendiente y la ordenada al origen son parametros de la función

h0(X) = Θ0 + Θ1x

El residuo se calcula con la diferencia entre el valor real y el valor de su prediccion.

Como este residuo puede ser negativo tomamos el cuadrado de estos.

La suma de los residuos al cuadrado representan la funcion a minimizar 

RSS

## Run and evaluate the function

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()

# Run the regresion
regression.fit(X, y) # Fit or "Train" the model
```

### Slope Coefficient

```python
regression.coef_ # Theta 1
```

Como la pendiente es positiva indica que existe una relación positiva entre las variables. En este caso mientras mayor sea el presupuesto, mejor será la película

### Intercept

```python
regression.intercept_
```

"-7236192" no parece tener sentido al hacer una película con 0 budget. Esto pasa porque la regresión lineal es una gran simplificación de la realidad.

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)

plt.plot(X, regression.predict(X), color="red", linewidth=3)

plt.title("Film cost vs Film Revenue")
plt.xlabel("Production Budget")
plt.ylabel("Revenue U$D")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

¿Que tanto representa nuestro modelo a la realidad?

R^2; "The goodness of fit"

```python
regression.score(X,y) # Is the amount of variation in film revenue explained by film budget
```

La variable de presupuesto puede explicar el 54% de la varianza de la ganancia de las películas
