## Example 1 - A simple cost function

### $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x+ 1
```

```python
# Make data
x_1 = np.linspace(start = -3, stop = 3, num = 100)
```

```python
# Plot
plt.xlim([-3,3])
plt.ylim([0,8])

plt.xlabel("x", fontsize="16")
plt.ylabel("f(x)", fontsize="16")

plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and Derivates
Challenge: Create a python function for the derivative of $f(x)$ called $df(x)$

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot function and derivative side by side
plt.style.use("dark_background")

plt.figure(figsize=[20,7])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.grid()

plt.xlim([-3,3])
plt.ylim([0,8])

plt.xlabel("x", fontsize="16")
plt.ylabel("f(x)", fontsize="16")

plt.title("Cost function", fontsize=25)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)

# 2 Chart: CDerivative
plt.subplot(1, 2, 2)
plt.grid()
plt.title("Slope of the cost function", fontsize=25)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=3)

plt.xlabel("x", fontsize="16")
plt.ylabel("df(x)", fontsize="16")

plt.show()
```

## Python Loops & Gradient Descent

```python
# Python For Loop
for counter in range(5):
    print("Hola mundo", counter)

print("Aca termina el loop")
```

```python
# Python while loop
counter = 0
while counter < 7:
    print("Contando . . . ", counter)
    counter = counter + 1
print("Done")
```

```python
# Gradient Descent

new_x = 3
previous_x = 0
step_multiplier = 0.1

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

print("Local minimum occurs at: ", new_x)
print("Slope or df(x) value at this point is: ", df(new_x))
print("f(x) value or cost at this point is: ", f(new_x))
```

```python
new_x = 2
previous_x = 0
step_multiplier = 0.1
precision= 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if(step_size < precision):
        print("Loop ran " , n , " times")
        break

print("Local minimum occurs at: ", new_x)
print("Slope or df(x) value at this point is: ", df(new_x))
print("f(x) value or cost at this point is: ", f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot
plt.style.use("dark_background")

plt.figure(figsize=[30,7])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)
plt.grid()

plt.xlim([-3,3])
plt.ylim([0,8])

plt.xlabel("x", fontsize="16")
plt.ylabel("f(x)", fontsize="16")

plt.title("Cost function", fontsize=25)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)

values = np.array(x_list)
plt.scatter(x_list, f(values), color="red", s=100, alpha = 0.7)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.title("Slope of the cost function", fontsize=25)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=3)
plt.scatter(x_list, slope_list, color="red",s=100, alpha=0.7)

plt.xlabel("x", fontsize="16")
plt.ylabel("df(x)", fontsize="16")


# 2 Chart: Derivative ( Close Up)
plt.subplot(1, 3, 3)


plt.title("Gradient Descent (Close up)", fontsize=25)
plt.xlabel("x", fontsize="16")
plt.grid()
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)

plt.plot(x_1, df(x_1), color="skyblue", linewidth=6, alpha=0.7)
plt.scatter(x_list, slope_list, color="red",s=180, alpha=0.7)

plt.show()
```
