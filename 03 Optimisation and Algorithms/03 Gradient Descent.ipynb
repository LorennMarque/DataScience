## Example 1 - A simple cost function

### $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x+ 1
```

## Slope and Derivates
Challenge: Create a python function for the derivative of $f(x)$ called $df(x)$

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[20,7])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.grid()

plt.xlim([-3,3])
plt.ylim([0,8])

plt.xlabel("x", fontsize="16")
plt.ylabel("f(x)", fontsize="16")

plt.title("Cost function", fontsize=25)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)

# 2 Chart: CDerivative
plt.subplot(1, 2, 2)
plt.grid()
plt.title("Slope of the cost function", fontsize=25)
plt.plot(x_1, df(x_1), color="skyblue", linewidth=3)

plt.xlabel("x", fontsize="16")
plt.ylabel("df(x)", fontsize="16")

plt.show()
```

## Example 2 - Multiple Minima initial Guess & Advanced Functions

### $ g(x) = x^4 - 4x^2 + 5 $

```python
# Initial Data
x_2 = np.linspace(-2, 2, 1000)

# Challenge: Write the g(x) function and the dg(x) function in Python
def g(x):
    return x ** 4 - 4 * x ** 2 + 5
```

```python
def dg(x):
    return 4 * x ** 3 - 8 * x
```

```python
# Plot function and derivative side by side
plt.style.use("dark_background")

plt.figure(figsize=[20,7])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.grid()

plt.xlim([-2,2])
plt.ylim([0.5,5.5])

plt.xlabel("x", fontsize="16")
plt.ylabel("g(x)", fontsize="16")

plt.title("Cost function", fontsize=25)
plt.plot(x_2, g(x_2), color="blue", linewidth=3)

# 2 Chart: CDerivative
plt.subplot(1, 2, 2)
plt.grid()
plt.title("Slope of the cost function", fontsize=25)

plt.plot(x_2, dg(x_2), color="skyblue", linewidth=3)

plt.xlim([-2,2])
plt.ylim([-6,8])

plt.xlabel("x", fontsize="16")
plt.ylabel("dg(x)", fontsize="16")

plt.show()
```

## Gradient Descent as a Python Function

```python
def gradient_descent(derivative_func, initial_guess, multiplier = 0.02, precision = 0.001, max_iter=300):
    new_x = initial_guess

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if(step_size < precision):
            print("Loop ran " , n , " times")
            break

    return new_x, x_list, slope_list
```

```python
# local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = 0.5, multiplier = 0.02, precision =  0.001)
local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = 0.5)

print('Local min occuts at: ', local_min)
print('Number of steps:', len(list_x))
```

```python
# Calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = -0.1)

# Plot function and derivative and scatter plot side by side
plt.figure(figsize=[20,7])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.grid()

plt.xlim([-2,2])
plt.ylim([0.5,5.5])

plt.xlabel("x", fontsize="16")
plt.ylabel("g(x)", fontsize="16")

plt.title("Cost function", fontsize=25)
plt.plot(x_2, g(x_2), color="blue", linewidth=3, alpha=0.8)
plt.scatter(list_x, g(np.array(list_x)), color="red", s=100, alpha=0.6)

# 2 Chart: CDerivative
plt.subplot(1, 2, 2)
plt.grid()
plt.title("Slope of the cost function", fontsize=25)

plt.plot(x_2, dg(x_2), color="skyblue", linewidth=3)
plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.6)

plt.xlim([-2,2])
plt.ylim([-6,8])

plt.xlabel("x", fontsize="16")
plt.ylabel("dg(x)", fontsize="16")

plt.show()
```

## Example 3 - Divergence, Overflos, and Python Tuples
### $ h(x) = x^5 - 2x^4 + 2$

```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=1000)

def h(x):
    return x**5 - 2 * x ** 4 + 2

def dh(x):
    return 5 * x ** 4 - 8 * x ** 3
```

```python
# Calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(derivative_func = dh, initial_guess = -0.2, max_iter=70)

# Plot function and derivative and scatter plot side by side
plt.figure(figsize=[20,7])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.grid()

plt.xlim([-1.2,2.5])
plt.ylim([-1, 4])

plt.xlabel("x", fontsize="16")
plt.ylabel("h(x)", fontsize="16")

plt.title("Cost function", fontsize=25)
plt.plot(x_3, h(x_3), color="blue", linewidth=3, alpha=0.8)
plt.scatter(list_x, h(np.array(list_x)), color="red", s=100, alpha=0.6)

# 2 Chart: CDerivative
plt.subplot(1, 2, 2)
plt.grid()
plt.title("Slope of the cost function", fontsize=25)

plt.plot(x_3, dh(x_3), color="skyblue", linewidth=3)
plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.6)

plt.xlim([-1,2])
plt.ylim([-4,5])

plt.xlabel("x", fontsize="16")
plt.ylabel("dg(x)", fontsize="16")

plt.show()

print("Local min occurs at: ", local_min)
print("Cost at this minimum is: ", h(local_min))
print("Number of steps: ", len(list_x))

```
